```python
import cv2
import matplotlib.pyplot as plt

img_no1 = cv2.imread('001.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_no1, cv2.COLOR_BGR2RGB))
plt.show()

img_no1 = cv2.cvtColor(img_no1, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_no1, cv2.COLOR_GRAY2RGB))
plt.show()
```
